## Numerical Variables

```python
df_final.columns
```

```python
rating = df_final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
rating = rating.rename(columns={0:'Rating value'})
rating
```

## The Rating says

1. 0.0 give white   --> not rated
2. 1.8-2.4 gives red   --> Poor
3. 2.5-3.4 gives orange   --> Average
4. 3.5-3.9 gives yellow   --> Good
5. 4.0-4.4 gives green   --> Very good
6. 4.5-4.9 gives dark green   -->Excellent


```python
plt.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating", y="Rating value", data=rating)
```

```python

sns.barplot(x="Aggregate rating", y="Rating value",hue='Rating color', data=rating, palette=['gray','red','orange','yellow','green','blue'])
```

Observation: 

1. Maximum ratings are b/w 2.5 to 3.4.
2. Count of Not rated is very high

```python
# count plot, it only shows the dataset's count, don't show the actual count of the ratings

sns.countplot(x="Rating color", data = rating, palette=['gray','red','orange','yellow','green','blue'], width=0.6)
```
